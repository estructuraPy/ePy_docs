# Anexos {.unnumbered}

## Anexo 1.1: Ejecución de pruebas de vuelco

![](images/prueba_vuelco.png)

La ejecución de la prueba de vuelco se puede realizar siguiente el siguiente proceso:

1.  Se coloca un contenedor de dimensiones definidas el cual debe estar construido por un material capaz de almacenar y resistir el peso del material. Dicho contenedor debe estar dividido horizontalmente de manera que permita el desplazamiento del material cuando sea inclinado.
2.  Se debe llenar el contenedor con el material por analizar. En caso de requerirlo, dicho material puede ser compactado. Para determinar la densidad y la masa del material se pueden hacer mediciones de densidad in situ y relacionar con el volumen del contenedor.
3.  Despejar los alrededores del contenedor de manera que el material sobrante no influya al momento de levantar el contenedor desde unos de sus lados
4.  Se inicia el izaje desde uno de los lados del contenedor asegurando que el lado opuesto quede como punto de apoyo y no sufra desplazamientos.
5.  Registre el ángulo al cual se da el desplazamiento de la parte superior del contenedor. Dicho ángulo será el ángulo a. El esfuerzo normal será obtenido con el peso de la parte superior desplazada y las dimensiones del contenedor.
