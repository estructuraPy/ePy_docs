## 1ï¸âƒ£ Imports y ConfiguraciÃ³n

````python
# Imports necesarios
import pandas as pd
import numpy as np
from pathlib import Path

# API PÃºblica de ePy_docs
from ePy_docs.writers import ReportWriter

print("âœ… ePy_docs importado correctamente")
print(f"ğŸ“¦ VersiÃ³n de pandas: {pd.__version__}")
print(f"ğŸ“¦ VersiÃ³n de numpy: {np.__version__}")
````

## 2ï¸âƒ£ Crear Escritor de Documentos

````python
# Crear el escritor de reportes
writer = ReportWriter()

print("âœ… ReportWriter inicializado")
print(f"ğŸ“Š Contador de tablas: {writer.table_counter}")
print(f"ğŸ–¼ï¸  Contador de figuras: {writer.figure_counter}")
print(f"ğŸ“ Buffer vacÃ­o: {len(writer.get_content()) == 0}")
````

## 3ï¸âƒ£ TÃ­tulo y Estructura del Documento

````python
# Construir estructura del documento con method chaining
(writer
    .add_h1("Reporte TÃ©cnico de AnÃ¡lisis de Datos")
    .add_content("**Autor:** Equipo de IngenierÃ­a")
    .add_content("**Fecha:** Octubre 2025")
    .add_content("**VersiÃ³n:** 1.0.0")
    .add_content("\n---\n")
)

print("âœ… TÃ­tulo y metadata agregados")
print(f"ğŸ“„ Contenido actual: {len(writer.get_content())} caracteres")
````

## 4ï¸âƒ£ SecciÃ³n de IntroducciÃ³n

````python
# Agregar secciÃ³n de introducciÃ³n
(writer
    .add_h2("1. IntroducciÃ³n")
    .add_content("""
Este reporte demuestra las capacidades completas de la biblioteca **ePy_docs** 
para la generaciÃ³n automatizada de documentos tÃ©cnicos profesionales.

El sistema permite crear reportes con contenido enriquecido incluyendo:
- Tablas de datos
- Ecuaciones matemÃ¡ticas
- ImÃ¡genes y grÃ¡ficos
- CÃ³digo fuente
- Referencias bibliogrÃ¡ficas
""")
    .add_callout("Este documento fue generado automÃ¡ticamente usando solo la API pÃºblica de ePy_docs.", type="note")
)

print("âœ… IntroducciÃ³n agregada")
print(f"ğŸ“ Total de callouts: {writer.note_counter}")
````

## 5ï¸âƒ£ Tablas de Datos con Pandas

````python
# Crear datos de ejemplo
np.random.seed(42)
data = {
    "Componente": ["Columna-A", "Columna-B", "Viga-1", "Viga-2", "Losa"],
    "Carga Axial (kN)": [350.5, 420.8, 125.3, 145.7, 85.2],
    "Momento (kNÂ·m)": [45.2, 52.8, 28.5, 32.1, 15.4],
    "Cortante (kN)": [18.5, 22.3, 35.8, 38.2, 12.7],
    "Estado": ["OK", "OK", "OK", "Revisar", "OK"]
}

df = pd.DataFrame(data)

# Agregar tabla al documento
(writer
    .add_h2("2. AnÃ¡lisis de Cargas Estructurales")
    .add_content("La siguiente tabla muestra los resultados del anÃ¡lisis de cargas para los elementos estructurales principales:")
    .add_table(df, title="Resultados de AnÃ¡lisis Estructural")
    .add_content("\nComo se observa en la tabla anterior, todos los elementos cumplen con los requisitos de diseÃ±o excepto la Viga-2 que requiere revisiÃ³n.")
)

print("âœ… Tabla agregada")
print(f"ğŸ“Š Contador de tablas: {writer.table_counter}")
df.head()
````

## 6ï¸âƒ£ Ecuaciones MatemÃ¡ticas

````python
# Agregar ecuaciones matemÃ¡ticas
(writer
    .add_h2("3. Fundamentos TeÃ³ricos")
    .add_content("La capacidad de carga de una columna se calcula segÃºn la siguiente ecuaciÃ³n:")
    .add_equation(
        r"P_{n} = 0.85 f'_{c} (A_{g} - A_{st}) + f_{y} A_{st}",
        label="eq-capacidad",
        caption="Capacidad nominal de columna con confinamiento"
    )
    .add_content(r"donde $f'_c$ es la resistencia del concreto y $f_y$ es el esfuerzo de fluencia del acero.")
    .add_content("\nLa relaciÃ³n esbeltez-capacidad estÃ¡ dada por:")
    .add_equation(
        r"\lambda = \frac{KL}{r} \leq \lambda_{max}",
        caption="RelaciÃ³n de esbeltez"
    )
    .add_callout(r"Para columnas cortas, tÃ­picamente $\lambda < 40$ para elementos sin arriostres laterales.", type="tip")
)

print("âœ… Ecuaciones agregadas")
print(f"ğŸ“ Contenido matemÃ¡tico incluido")
````

## 7ï¸âƒ£ ImÃ¡genes y Figuras

````python
# Agregar imagen existente
image_path = "data/user/document/03_geotech/images/enrocado_esquema.png"

(writer
    .add_h2("4. Esquemas de DiseÃ±o")
    .add_content("El siguiente esquema ilustra la configuraciÃ³n tÃ­pica de un enrocado de protecciÃ³n:")
    .add_image(image_path, caption="Esquema tÃ­pico de enrocado", width="80%")
    .add_callout("La geometrÃ­a del enrocado debe considerar la altura de ola de diseÃ±o y el Ã¡ngulo de talud.", type="important")
)

print(f"âœ… Imagen agregada: {Path(image_path).name}")
print(f"ğŸ–¼ï¸  Contador de figuras: {writer.figure_counter}")
````

## 8ï¸âƒ£ CÃ³digo Ejecutable

````python
# Agregar cÃ³digo de ejemplo
code_example = """
import numpy as np

def calcular_capacidad(fc, fy, Ag, Ast):
    '''Calcula la capacidad nominal de una columna'''
    P_concreto = 0.85 * fc * (Ag - Ast)
    P_acero = fy * Ast
    return P_concreto + P_acero

# Ejemplo de uso
fc = 28  # MPa
fy = 420  # MPa
Ag = 0.16  # mÂ²
Ast = 0.008  # mÂ²

P_n = calcular_capacidad(fc, fy, Ag, Ast)
print(f"Capacidad nominal: {P_n:.2f} kN")
"""

(writer
    .add_h2("5. ImplementaciÃ³n Computacional")
    .add_content("El siguiente cÃ³digo Python implementa el cÃ¡lculo de capacidad segÃºn la ecuaciÃ³n presentada:")
    .add_chunk(code_example, language="python")
    .add_content("\nEste algoritmo puede ser integrado en sistemas de diseÃ±o automatizado.")
)

print("âœ… CÃ³digo agregado")
print(f"ğŸ’» Chunk de cÃ³digo incluido")
````

## 9ï¸âƒ£ Listas y Callouts

````python
# Agregar listas y callouts
requisitos = [
    "Resistencia del concreto mÃ­nima de 28 MPa",
    "Acero de refuerzo grado 420",
    "Recubrimiento mÃ­nimo de 40 mm",
    "RelaciÃ³n de esbeltez menor a 40"
]

pasos_diseno = [
    "Determinar cargas de diseÃ±o",
    "Calcular combinaciones de carga",
    "Verificar capacidad a compresiÃ³n",
    "Revisar momentos de flexiÃ³n",
    "Validar cortante",
    "DiseÃ±ar refuerzo transversal"
]

(writer
    .add_h2("6. Criterios de DiseÃ±o")
    .add_h3("6.1 Requisitos Normativos")
    .add_content("Los elementos estructurales deben cumplir con los siguientes requisitos:")
    .add_unordered_list(requisitos)
    .add_callout("El incumplimiento de cualquiera de estos requisitos puede comprometer la seguridad estructural.", type="warning")
    .add_h3("6.2 Procedimiento de DiseÃ±o")
    .add_content("El proceso de diseÃ±o sigue los siguientes pasos:")
    .add_ordered_list(pasos_diseno)
    .add_callout("Cada paso debe ser verificado por un ingeniero estructural certificado.", type="caution")
)

print("âœ… Listas y callouts agregados")
print(f"ğŸ“‹ Lista no ordenada: {len(requisitos)} items")
print(f"ğŸ“‹ Lista ordenada: {len(pasos_diseno)} pasos")
````

## ğŸ¯ GeneraciÃ³n de Documentos (HTML + PDF)

````python
# Generar HTML y PDF simultÃ¡neamente
print("ğŸ”„ Generando HTML y PDF...")
print("   (Esto puede tomar algunos segundos...)\n")

result = writer.generate(html=True, pdf=True, output_filename="demo_api_publica")

# Verificar HTML
if result.get('html'):
    from pathlib import Path
    html_path = Path(result['html'])
    if html_path.exists():
        html_size = html_path.stat().st_size / 1024
        print(f"âœ… HTML generado: {html_path.name}")
        print(f"   ğŸ’¾ TamaÃ±o: {html_size:.1f} KB")
    else:
        print(f"âš ï¸  HTML: Ruta devuelta pero archivo no encontrado")
else:
    print("âŒ HTML no generado")

# Verificar PDF
if result.get('pdf'):
    pdf_path = Path(result['pdf'])
    if pdf_path.exists():
        pdf_size = pdf_path.stat().st_size / 1024
        print(f"\nâœ… PDF generado: {pdf_path.name}")
        print(f"   ğŸ’¾ TamaÃ±o: {pdf_size:.1f} KB")
    else:
        print(f"\nâš ï¸  PDF: Ruta devuelta pero archivo no encontrado")
else:
    print("\nâŒ PDF no generado")

# EstadÃ­sticas del documento
print(f"\nğŸ“Š EstadÃ­sticas del Documento:")
print(f"   â€¢ Tablas: {writer.table_counter}")
print(f"   â€¢ Figuras: {writer.figure_counter}")
print(f"   â€¢ Callouts: {writer.note_counter}")
````
