## üé® Generador de Cat√°logo de Paletas

```python
def load_color_data():
    """Cargar datos de colores originales y optimizados"""
    
    config_dir = Path("src/ePy_docs/config")
    
    # Cargar colores originales
    colors_file = config_dir / "colors.epyson"
    with open(colors_file, 'r', encoding='utf-8') as f:
        original_colors = json.load(f)
    
    # Cargar colores optimizados si existen
    optimized_file = config_dir / "colors_optimized.epyson"
    optimized_colors = {}
    if optimized_file.exists():
        with open(optimized_file, 'r', encoding='utf-8') as f:
            optimized_colors = json.load(f)
    
    return original_colors, optimized_colors

def visualize_palette(palette_name, palette_data, optimized_data=None):
    """Visualizar una paleta de colores"""
    
    fig, ax = plt.subplots(1, 1, figsize=(12, 3))
    
    # Colores principales a mostrar
    main_colors = ['primary', 'secondary', 'tertiary', 'quaternary', 'page_background']
    colors_to_show = []
    labels = []
    
    for color_key in main_colors:
        if color_key in palette_data:
            color = palette_data[color_key]
            if isinstance(color, list) and len(color) == 3:
                # Convertir RGB a hex
                hex_color = f"#{color[0]:02x}{color[1]:02x}{color[2]:02x}"
                colors_to_show.append(hex_color)
                labels.append(color_key.title())
    
    # Crear visualizaci√≥n
    for i, (color, label) in enumerate(zip(colors_to_show, labels)):
        rect = patches.Rectangle((i, 0), 1, 1, facecolor=color, edgecolor='black', linewidth=1)
        ax.add_patch(rect)
        ax.text(i + 0.5, 0.5, label, ha='center', va='center', fontweight='bold', 
                color='white' if sum(int(color[j:j+2], 16) for j in (1, 3, 5)) < 384 else 'black')
    
    ax.set_xlim(0, len(colors_to_show))
    ax.set_ylim(0, 1)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # T√≠tulo con informaci√≥n de optimizaci√≥n
    optimization_status = "‚ú® Optimizada" if optimized_data and any(k.endswith('_high_contrast') for k in optimized_data.keys()) else "üìã Original"
    plt.title(f"üé® Paleta: {palette_name.title()} {optimization_status}", fontsize=14, fontweight='bold', pad=20)
    
    plt.tight_layout()
    plt.show()

# Cargar y visualizar paletas
print("üé® CAT√ÅLOGO DE PALETAS DE COLORES")
print("=" * 50)

original_colors, optimized_colors = load_color_data()
palettes = original_colors.get('palettes', {})
optimized_palettes = optimized_colors.get('palettes', {})

print(f"üìä Total paletas: {len(palettes)}")
print(f"‚ú® Paletas optimizadas: {len([p for p in optimized_palettes.values() if isinstance(p, dict) and any(k.endswith('_high_contrast') for k in p.keys())])}")
print()

# Mostrar algunas paletas principales
main_palettes = ['blues', 'greens', 'classic', 'professional', 'corporate']
for palette_name in main_palettes:
    if palette_name in palettes:
        palette_data = palettes[palette_name]
        optimized_data = optimized_palettes.get(palette_name, {})
        if isinstance(palette_data, dict):
            visualize_palette(palette_name, palette_data, optimized_data)
```

```python
def demonstrate_chunk_differentiation():
    """Demostrar la diferenciaci√≥n de chunks en el sistema"""
    
    print("üìÑ DIFERENCIACI√ìN DE CHUNKS EN PDF")
    print("=" * 50)
    print()
    
    # Informaci√≥n sobre la implementaci√≥n
    chunk_info = {
        'display': {
            'name': 'Display Chunks',
            'color': 'Fondo gris claro (gray!5)',
            'border': 'Borde izquierdo gris (2pt)',
            'uso': 'C√≥digo de ejemplo, sintaxis, configuraci√≥n',
            'ejemplo': 'chunk_type="display"'
        },
        'executable': {
            'name': 'Executable Chunks',
            'color': 'Fondo naranja/caf√© (orange!10)',
            'border': 'Borde izquierdo naranja (2pt)',
            'uso': 'C√≥digo que se ejecuta al renderizar',
            'ejemplo': 'chunk_type="executable"'
        }
    }
    
    print("üé® TIPOS DE CHUNKS:")
    print()
    
    for chunk_type, info in chunk_info.items():
        print(f"üì¶ **{info['name']}**")
        print(f"   üé® Color: {info['color']}")
        print(f"   üî≤ Borde: {info['border']}")
        print(f"   üí° Uso: {info['uso']}")
        print(f"   üìù Sintaxis: {info['ejemplo']}")
        print()
    
    print("üîß IMPLEMENTACI√ìN:")
    print("   ‚úÖ Paquete: mdframed (compatible con XeTeX)")
    print("   ‚úÖ Definiciones: Din√°micas desde layouts")
    print("   ‚úÖ Fallback: Sistema robusto de respaldo")
    print("   ‚úÖ Font Detection: Evita fuentes problem√°ticas")
    print()
    
    print("üìä BENEFICIOS:")
    print("   ‚úì Consistencia HTML-PDF")
    print("   ‚úì Diferenciaci√≥n visual clara")
    print("   ‚úì WCAG AA compliant")
    print("   ‚úì Sin errores de compilaci√≥n")
    print()
    
    # Ejemplo de c√≥digo
    code_example = '''
# Ejemplo de uso:
from ePy_docs.writers import DocumentWriter

writer = DocumentWriter(document_type="report", layout_style="minimal")

# Display chunk (fondo claro)
writer.add_code_chunk(
    code='print("Ejemplo de c√≥digo")',
    language="python",
    chunk_type="display",
    caption="C√≥digo de ejemplo"
)

# Executable chunk (fondo naranja)
writer.add_code_chunk(
    code='print("C√≥digo ejecutable")',
    language="python",
    chunk_type="executable",
    caption="C√≥digo que se ejecuta"
)

# Generar
writer.generate(pdf=True, html=True, output_filename="documento")
'''
    
    print("üíª EJEMPLO DE USO:")
    print(code_example)
    
    # Archivos relacionados
    print("\nüìÅ ARCHIVOS CLAVE:")
    print("   ‚Ä¢ src/ePy_docs/core/_pdf.py - Generaci√≥n LaTeX")
    print("   ‚Ä¢ src/ePy_docs/core/_code.py - Formato de chunks")
    print("   ‚Ä¢ src/ePy_docs/core/_text.py - API add_code_chunk")
    print("   ‚Ä¢ src/ePy_docs/writers.py - API p√∫blica")
    
    return True

# Ejecutar demostraci√≥n
success = demonstrate_chunk_differentiation()

if success:
    print("\n‚úÖ Demostraci√≥n completada")
    print("üöÄ Sistema listo para generar PDFs con diferenciaci√≥n visual")
```
