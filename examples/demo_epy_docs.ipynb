## üíª 9. Bloques de C√≥digo

Inclusi√≥n de c√≥digo Python con resaltado de sintaxis:

```python
writer.add_h2("7. C√≥digo de An√°lisis")

writer.add_h3("7.1 Script de Procesamiento de Datos")

codigo_ejemplo = '''
import numpy as np
import pandas as pd

def calcular_momento_resistente(As, fy, d, a):
    """
    Calcula el momento resistente nominal de una secci√≥n.
    
    Args:
        As: √Årea de acero (cm¬≤)
        fy: Esfuerzo de fluencia (kg/cm¬≤)
        d: Peralte efectivo (cm)
        a: Profundidad del bloque de compresi√≥n (cm)
    
    Returns:
        Momento resistente (kN¬∑m)
    """
    Mn = As * fy * (d - a/2) / 100000  # Convertir a kN¬∑m
    return Mn

# Ejemplo de uso
As = 25.13  # cm¬≤
fy = 4200   # kg/cm¬≤
d = 45      # cm
a = 8       # cm

Mn = calcular_momento_resistente(As, fy, d, a)
print(f"Momento resistente: {Mn:.2f} kN¬∑m")
'''

writer.add_code_chunk(
    codigo_ejemplo,
    language="python",
    chunk_type="display",
    caption="Funci√≥n para c√°lculo de momento resistente"
)

print("‚úÖ Bloques de c√≥digo agregados")
```
